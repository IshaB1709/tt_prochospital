```python
claims_df=spark.read.parquet("/mnt/bronze/claims/*",header= True)
```

```python
import pyspark.sql.functions as f

claims_df=claims_df.withColumn("data_source", f.when(f.input_file_name().contains("hospital1"), "hosa")
.when(f.input_file_name().contains("hospital2"), "hosb")
.otherwise(None))
```

```python
claims_df.filter(claims_df.data_source=="hosb").display()
```

```python
claims_df.write.mode("overwrite").format("parquet").save("/mnt/silver/claims")
```
